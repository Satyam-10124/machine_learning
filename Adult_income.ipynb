```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('./datas/adult.csv')
df.head()
```

```python
for col in df:
    df[col] = df[col].replace("?",df[col].mode()[0])
df
```

```python
df.workclass.value_counts()
```

```python
df.workclass = df.workclass.replace(["Without-pay","Never-worked"],"others")
df.workclass = df.workclass.replace(["Local-gov","State-gov",'Federal-gov'],"gov")
df.workclass = df.workclass.replace(["Self-emp-not-inc","Self-emp-inc"],"Self-emp")
df.workclass.unique()
```

```python
df.education.value_counts()
```

```python
df.education = df.education.replace(["HS-grad","11th",'10th','7th-8th','9th','12th','5th-6th','1st-4th','Preschool'],"school")
df.education.value_counts()
```

```python
df.occupation.value_counts()
```

```python
df['race'].replace(['Black', 'Asian-Pac-Islander', 'Amer-Indian-Eskimo', 'Other'],' Other', inplace = True)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder();
```

```python
cat_variable = df.select_dtypes(include = "object").columns
for col in cat_variable:
    df[col] = le.fit_transform(df[col])
df
```

```python
cor = df.corr()
cor
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df[['age','workclass','fnlwgt','education','education.num','occupation','relationship','race','sex','capital.gain','hours.per.week','native.country']],df.income, test_size = 0.3, random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler() 
x_train = pd.DataFrame(scaler.fit_transform(x_train), columns = x_train.columns)
x_test = pd.DataFrame(scaler.transform(x_test), columns = x_test.columns)
```
