```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
import warnings
```

```python
df = pd.read_csv('./datas/HR_comma_sep.csv')
le = LabelEncoder()
df.loc[:, 'salary_le'] = le.fit_transform(df.salary)
df.loc[:, 'department_le'] = le.fit_transform(df.Department)
df.head(10)
```

```python
df.describe().T
```

```python
x_train, x_test, y_train, y_test = train_test_split(df[['satisfaction_level','average_montly_hours','average_montly_hours','salary_le','department_le']], df.left, test_size = 0.2)
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test, y_test)
```
